```python
# Import Dependencies

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Read the CSV File

file = 'crypto_data.csv'
df = pd.read_csv(file)
df
```

```python
df.dtypes
```

```python
# Remove all currencies that are not being traded. 

traded_true = df[df.IsTrading]
traded_true
```

```python
# Drop the IsTrading column

traded = traded_true.drop('IsTrading', 1)
traded
```

```python
# Remove all rows that have at least one null value

traded_drop_null = traded.dropna(how='any',axis=0) 
traded_drop_null
```

```python
# Filter for cryptocurrencies that are being mined (total coins > 0)

greater_70 = traded_drop_null[traded_drop_null['TotalCoinsMined'] > 0] 
greater_70
```

```python
# Drop CoinName

data = greater_70.drop('CoinName', 1)
data
```

```python
# Renamed Unnamed: 0 to "Ticker"

df = data.rename(columns={'Unnamed: 0': 'Ticker'})
df
```

```python
# Convert Algorithm and ProofType to numeric data

new_df = pd.get_dummies(df, prefix=['Algorithm', 'ProofType'], columns=['Algorithm', 'ProofType'])
new_df
```

```python
final_df = new_df.drop('Ticker', 1)
```

```python
# Scale the data

# Create a scaler to standardize the data
scaler = StandardScaler().fit(final_df)

# Train the scaler with the X_train data.
scaled_crypto = scaler.transform(final_df)
scaled_crypto
```

```python
# PCA Model to discover the explained variance. We want to preserve 90% of the data. 

```

```python
# Initialize PCA model
pca = PCA(n_components=.90)

# Get principal components for the data.
crypto_pca = pca.fit_transform(scaled_crypto)
crypto_pca
```

```python
# Transform PCA data to a DataFrame
df_crypto_pca = pd.DataFrame(
    data=crypto_pca,
)
df_crypto_pca.head()
```

# Prior to the PCA Model --> there were 98 columns and 532 rows. 
# After preserving 90% of the data, we brought it down to 74 columns and 5 rows. 

```python
# Fetch the explained variance
pca.explained_variance_ratio_
```
